```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
# Create and Generate a Word Cloud Image
f = open("textdata.txt", "r")
text=f.read()
f.close()
```

```python
# Generate a word cloud image
stopwords = set(STOPWORDS)
mask = np.array(Image.open("Flag.png"))
wordcloud = WordCloud(stopwords=stopwords,background_color='white', max_words=1000, mask=mask,contour_color='#023075',contour_width=1,width=800, height=400, colormap='rainbow').generate(text)
# create image as cloud
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
# store to file
plt.savefig("cloud.png", format="png")
plt.show()
```
